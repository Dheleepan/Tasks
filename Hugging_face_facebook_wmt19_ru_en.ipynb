```python
from transformers import FSMTForConditionalGeneration, FSMTTokenizer
import pandas as pd
```

```python
!gdown --id 1FbWuwELO_ssoJ5nbm8fqOXcyzsqrUwX4
```

```python
mname = "facebook/wmt19-ru-en"
tokenizer = FSMTTokenizer.from_pretrained(mname)
model = FSMTForConditionalGeneration.from_pretrained(mname)
```

```python
df = pd.read_csv('/content/news.csv')
```

```python
count_russian_news = 21673
```

```python
actual_russian_news = []
translated_russian_news = []
```

```python
for i in range(count_russian_news):
  text = df.iloc[i]['text']
  actual_russian_news.append(text)
  input_ids = tokenizer.encode(text, return_tensors="pt")
  outputs = model.generate(input_ids)
  decoded = tokenizer.decode(outputs[0], skip_special_tokens=True)
  translated_russian_news.append(decoded)


```

```python
minimum_length_list = min(len(actual_russian_news),len(translated_russian_news))
```

```python
summerized_file = open('Translated.txt','w')
```

```python
for i in range(minimum_length_list):
  summerized_file = open("Translated.txt", "a")
  summerized_file.write('Actual_text\n')
  summerized_file.write('\n')
  summerized_file.write('\n')
  summerized_file.write(actual_russian_news[i]+'\n')
  summerized_file.write('\n')
  summerized_file.write('\n')
  summerized_file.write('summerized_text')
  summerized_file.write('\n')
  summerized_file.write('\n')
  summerized_file.write(translated_russian_news[i]+ '\n')
  summerized_file.write('\n')
  summerized_file.write('\n')
  summerized_file.write('--------------------------------------------------------------------------------------')
  summerized_file.write('\n')
  summerized_file.close()
```
