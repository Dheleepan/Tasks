```python
from transformers import pipeline # importing the modules that are required
import os
```

```python
summarizer = pipeline("summarization", model="facebook/bart-large-cnn") # setting the pipeline
```

```python
length = 511 # count of total text files to be summerized
count = 1
actual_text = [] # creating a empty list for appending actual text
summerized = []  # Another empty list is create for appending summerized text
file_name = ''
```

```python
for _ in range(length):
  file_name = ''
  if len(str(count)) == 1:
    file_name ='00'+str(count) 
    print(file_name)
    final_file_name = '/content/sport'+'/'+ file_name +'.txt' 
    f = open(final_file_name)
    text = f.read()
    text_length = len(text.split())
    actual_text.append(text)
    result = summarizer(text, max_length = text_length, min_length=30, do_sample=False)
    summerized.append(result[0]['summary_text'])
    count += 1

  elif len(str(count)) == 2:
    file_name = '0'+str(count)
    print(file_name)
    final_file_name = '/content/sport'+'/'+ file_name + '.txt'
    f = open(final_file_name)
    text = f.read()
    text_length = len(text.split())
    actual_text.append(text)
    result = summarizer(text, max_length= text_length, min_length=30, do_sample=False)
    summerized.append(result[0]['summary_text'])
    count += 1

  else:
    file_name = str(count)
    print(file_name)
    final_file_name = '/content/sport'+'/'+ file_name + '.txt'
    f = open(final_file_name)
    text = f.read()
    text_length = len(text.split())
    actual_text.append(text)
    result = summarizer(text, max_length= text_length, min_length=30, do_sample=False)
    summerized.append(result[0]['summary_text'])
    count += 1
    
```

```python
minimum_length_list = min(len(actual_text),len(summerized))
```

```python
summerized_file = open('summeried.txt','w')
```

```python
for i in range(minimum_length_list):
  summerized_file = open("summeried.txt", "a")
  summerized_file.write('Actual_text\n')
  summerized_file.write('\n')
  summerized_file.write('\n')
  summerized_file.write(actual_text[i]+'\n')
  summerized_file.write('\n')
  summerized_file.write('\n')
  summerized_file.write('summerized_text')
  summerized_file.write('\n')
  summerized_file.write('\n')
  summerized_file.write(summerized[i]+ '\n')
  summerized_file.write('\n')
  summerized_file.write('\n')
  summerized_file.write('--------------------------------------------------------------------------------------')
  summerized_file.write('\n')
  summerized_file.close()
```
